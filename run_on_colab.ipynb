## Step 3: Train YOLOv8 Model
We will train the `yolov8s.pt` model for 50 epochs on the GPU.

```python
from ultralytics import YOLO
import os

# Load the model
model = YOLO('yolov8s.pt')

# Define training parameters
# Note: ensuring data.yaml path is correct relative to where we unzipped
data_path = '/content/dataset/data.yaml'

if not os.path.exists(data_path):
    # Attempt to find data.yaml if unzip structure is slightly different
    for root, dirs, files in os.walk('/content/dataset'):
        if 'data.yaml' in files:
            data_path = os.path.join(root, 'data.yaml')
            break
            
print(f"Training with config: {data_path}")

# Start training
results = model.train(
    data=data_path,
    epochs=50,
    imgsz=640,
    batch=16,
    project='ResQ-AI-Phase1',
    name='yolov8s_aider_colab',
    exist_ok=True,
    device=0
)
```
